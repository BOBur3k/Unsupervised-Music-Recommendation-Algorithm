Data visualization 

```python
sns.histplot(data = df, x = 'release_date')
plt.title('Number of songs released since 1950')
plt.xlabel('Release Year')
plt.ylabel('Number of songs released')
plt.show()
sns.histplot(data = group_release)
plt.title('Number of songs released in each decade since 1950s')
plt.xlabel('Decades')
plt.ylabel('Number of songs released')
plt.show()
#following visualizations shows how many songs were released in each year/decades
#based on this graphical representation we can observe that most amount of songs were released recently in the last decade 
#the second and third wave of song releses were during 1980s and 2000s 
```

Review of the Genre and Topics 

```python
sns.histplot(data = df, x = 'genre')
#based on this histplot, we can determine that we have more pop songs vs every other genre. The list popular genre is hip hop
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=df, x='genre', hue= group_release)
plt.show()
#based on the visualization below, we can make couple of takeaways: 
# 1 for the genre of POP music the rise of releases was 1980s and 2010s
# 2 for the genre of COUNTRY music the rise of releases was 1980s and 1960s 
# 3 for the genre of BLUES music the rise of releases was 2010s and 2000s 
# 4 for the genre of JAZZ music the rise of releases was 2010s and 2000s 
# 5 for the genre of REGGIE music the rise of releases was 2010s and 2000s 
# 6 for the genre of ROCK music the rise of releases was 2010s and 1980s 
# 7 for the genre of HIP HOP music the rise of releases was 2010s and 2000s  
```

```python
#What is the most popular topic/theme of our songs? 
topics = df['topic'].value_counts()
print("Top 5 of the topics popular among our dataset:")
print(topics.head().to_string(header=False))
print("\nLeast popular topics among our dataset:") 
print (topics.tail(3).to_string(header=False))

plt.figure(figsize=(10, 8))
sns.countplot(data=df, x= group_release, hue= 'topic')
plt.title('Number of songs based on the specific theme/topics released in each decade')
plt.xlabel('Release Years in decades')
plt.ylabel('Number of songs in the database')
plt.show()

#based on the findings below we can say that the most popular topic in our songs is sadness :(
#however we have a pick of popularity of Obsene and Violence theme songs in 2010s (society...)
```

```python
#well now when we know the trends for genres and trends for theme/topics of the songs. Time to see what is the common trend between genres of music and their main themes 
plt.figure(figsize=(10, 8))
sns.countplot(data = df, x = "genre", hue = 'topic')
plt.title('Genres of music and their dominant topics')
plt.xlabel('Genres of music')
plt.ylabel('Number of songs in the database')
plt.show()
#based on the visualizations below we can determine that POP and Country songs are mostly about sadness and wold/life 
#Blues is about Violence 
#Jazz about world/life 
#Raggie and Hip Hop about obsene
#Rock is about Violence 
```

```python
#last thing is to figure out how long is their songs, like how many creative wrting in each of the topics and genres? artists in what genre and topic wrtie more? 
plt.figure(figsize=(10, 8))
sns.barplot(data = df, x = 'genre', y = 'len')
plt.title('Number of lyrics in different genre of the songs')
plt.xlabel('Genres of songs')
plt.ylabel('Number of lyrics in the song')
plt.show()

plt.figure(figsize=(10, 8))
sns.barplot(data = df, x = 'topic', y = 'len')
plt.title('Number of lyrics in different topics of the songs')
plt.xlabel('Topics of songs')
plt.ylabel('Number of lyrics in the song')
plt.show()

#based on the visualization below we can say that Raggie and Hip Hop artists really dedicate their time and effort in writing lyrics in their songs
# Songs with obscene topic have the most lyrics than other songs 
```

Quantity of the songs 

```python
#I was wondering how many songs we have in our dataset and how many of them have a unique title. 
total_number_of_songs = len(df['track_name'])

print("Total number of songs:", total_number_of_songs)
print("Total number of UNIQUE songs:", df['track_name'].nunique())
```

```python
#Then I was wondering how mnay songs have the same title since the difference between number of songs and unique songs is more than 4k
popular_names = df['track_name'].value_counts()
nunique_titles = len(popular_names[popular_names>1])
print("# of titles of the songs that repeats more than 1 time:", nunique_titles)
```

```python
# Time to visualize it with Bar plot of total number of songs vs. number of unique songs vs songs with the same title
plt.figure(figsize=(10, 8))
sns.barplot(x=['Total # Songs', 'Unique Titles', 'Not Unique Titles'], y=[total_number_of_songs, df['track_name'].nunique(), nunique_titles])
plt.ylabel('Number of songs')
plt.title('Total Number of Songs VS Number of Songs with Unique Title VS Songs with The Same Title')
plt.show()
```

Review of the Artists 

```python
#let's determine how many artistst do we have in our database + how many of them represented by more than 1 song in our database 
sum_of_artists = df['artist_name'].nunique()
artists = df['artist_name'].value_counts()

artists_of_1_more = len(artists[artists>1])

plt.figure(figsize=(10, 8))
sns.barplot(x = ['Total Number of Artists', 'Artists of more than 1 song'], y = [sum_of_artists, artists_of_1_more])
plt.title('Total Number of Artists VS Artist who have more than 1 song in our data')
plt.ylabel('Number of Artitst')
plt.show()

print("Number of artists:", sum_of_artists)
print("Number of artists with more than one song:", artists_of_1_more)
```

```python
#Since we have 3308 artist that have more than 1 song in our database, time to find the most popular artists, maybe we have bias towards some of them

A_list = df['artist_name'].value_counts().head(10)
print("Top 10 artists in our database based on the # of songs:")
print (A_list.to_string(header= False))
print(' ')
plt.figure(figsize=[14, 12])
sns.barplot(x=A_list.index, y=A_list.values)
plt.title("Top 10 artist on the dataset")
plt.xlabel('Artist Name')
plt.ylabel('Number of songs')
plt.show()
```

```python
#well based on our visualization and data findings, we can conclude that we have 5 artists with more than 100 songs, let's see how diverse their genre and topics in their songs 
#first we are creating new DF with only our top 10 artist 
A_list_df = df[df['artist_name'].isin(A_list.index)]
print("Number of songs our most popular artists have in our dataset:", A_list_df["artist_name"].value_counts().sum())
print("\nList of genres of songs that our artists perform \n", A_list_df["genre"].value_counts().to_string(header=False))
print("\nWhat our top artists sing about? \n", A_list_df["topic"].value_counts().to_string(header=False))
#now time for more visualizations 
plt.figure(figsize=[12, 10])
plt.title("Visual representation of what type of songs our Top Artists perform")
sns.countplot(data= A_list_df, x= "genre", hue = "topic")
plt.show
```
